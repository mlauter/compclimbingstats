```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(font_scale=2)
```

```python
import glob
import os
path = os.path.abspath('data/results')
print(path)
```

```python
filenames = glob.glob(path + '/*.csv')
```

```python
dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename, index_col=None, header=0))

df = pd.concat(dfs, ignore_index=True)
df.head()
```

```python
all_categories = {k: list(v)[0] for k, v in df.groupby('cat_id')['cat']}
```

```python
all_categories
```

```python
cat_filter = [1,2,5,6,15,16,17,18,19,20,42,45,79,80,81,82,83,84,286,287,288,289,290,291]
categories = {k:all_categories[k] for k in cat_filter}
categories
```

```python
filtered_df = df.loc[df['cat_id'].isin(categories.keys())]
filtered_df
```

```python
F_cats = [k for k in categories.keys() if ('WOMEN' in categories[k] or 'W O M E N' in categories[k] or 'female' in categories[k])]
```

```python
len(F_cats)
```

```python
len([k for k in categories.keys() if k not in F_cats])
```

```python
filtered_df['gender'] = filtered_df['cat_id'].apply(lambda c: 'F' if c in F_cats else 'M')
```

```python
filtered_df
```

```python
conds = [filtered_df.cat.str.contains('boulder'), filtered_df.cat.str.contains('lead'), filtered_df.cat.str.contains('combined')]
choices = ['B', 'L', 'C']
filtered_df['discipline'] = np.select(conds, choices, default='')
filtered_df
```

```python
conds = [
    filtered_df.cat.str.contains('youth A'),
    filtered_df.cat.str.contains('youth B'),
    filtered_df.cat.str.contains('juniors'),
]
choices = ['YA', 'YB', 'YJ']
filtered_df['age'] = np.select(conds, choices, default='A')
filtered_df
```

```python
filtered_df.loc[(filtered_df['gender'] == 'F') & (filtered_df['discipline'] == 'L') & (filtered_df['age'] == 'A')]
```

```python
filtered_df['fullname'] = filtered_df['firstname'].str.upper() + " " + filtered_df['lastname'].str.upper()
```

```python
most_wins_all_comps = filtered_df.loc[filtered_df['rank'] == 1].groupby(filtered_df['fullname']).size().reset_index(name='counts').sort_values(by=['counts'], ascending=False).reset_index()
most_wins_all_comps
```

```python
most_podiums_all_comps = filtered_df.loc[filtered_df['rank'].isin([1, 2, 3])].groupby(filtered_df['fullname']).size().reset_index(name='counts').sort_values(by=['counts'], ascending=False).reset_index()
most_podiums_all_comps.loc[most_podiums_all_comps['counts'] > 1]
```
